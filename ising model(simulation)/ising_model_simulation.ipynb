```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=50 #grid dimensions
spins=np.random.choice([-1,1],size=(N,N))
```

```python
# function for calculating energy change
def energy_change(i,j):
    J=1
    N=spins.shape[0]
    neighbours=(
        spins[(i+1)%N,j]+spins[(i-1)%N,j]+spins[i,(j+1)%N]+spins[i,(j-1)%N]
    )
    dE=2*spins[i,j]*neighbours
    return dE

    
```

```python
#metropolis algorithm for fliping is accepted or rejected
def metropolis(spins,T):
    N=spins.shape[0]
    for _ in range(N*N):
        i=np.random.randint(0,N)
        j=np.random.randint(0,N)
        dE=energy_change(i,j)
        if dE<0 or np.random.rand()<np.exp(-dE/T):
            spins[i,j]=-spins[i,j]
    return spins


```

```python
temp=[0.5,1.0,2.0,2.5,4.0] #temperatures
steps=10000 
final={}
for T in temp:
    spins=np.random.choice([-1,1],size=(N,N))
    for i in range(steps):
        if i%100==0:
            print(i," ",T) # just to know the progress
        spins=metropolis(spins,T)

    final[T]=spins.copy()
```

```python
plt.figure(figsize=(15, 3))

for idx, T in enumerate(temp):
    plt.subplot(1, 5, idx + 1)
    plt.imshow(final[T], cmap='binary')
    plt.title(f"T = {T}")
    plt.axis('off')

plt.suptitle("ising model at various temp", fontsize=14)
plt.show()
```

```python
T=0.5
spins=np.random.choice([-1,1],size=(N,N))
snapshots=[500,1500,3000,5000,10000]
states={}
for step in range(1,max(snapshots)+1):
    metropolis(spins,T)
    if step in snapshots:
        states[step]=spins.copy()


```

```python
plt.figure(figsize=(15, 3))

for idx, step in enumerate(snapshots):
    plt.subplot(1, 5, idx + 1)
    plt.imshow(states[step], cmap='binary')
    plt.title(f"Step {step}")
    
plt.show()
```

# Conclusion
at low temperatures ,the spin form large,uniform clustors where all spins are aligned in one direction.This shows strong ordering and low energy configurations<br>
at moderate temperatures,the system shows with partial order,mixed regions are present some of are aligned and some of are not.<br>
at high temperatures,the system was completely demagnetised and all are randomly aligned +1 or -1 due to thermal energy domination it destroys spin alignment.
