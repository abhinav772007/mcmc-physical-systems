```python
#part A
import numpy as np
V=np.array([420, 445, 380, 510, 398, 475, 425, 390, 460, 440])
avg_speed=np.average(V)
print("Average speed =",avg_speed,"m/s")
```

```python
#part B
std_speed=np.std(V)
print("Standard deviation =", std_speed,"m/s")
```

```python
#part c
import matplotlib.pyplot as plt
numbers=range(1,len(V)+1)
plt.bar(numbers,V)
plt.axhline(avg_speed)
plt.xlabel("Particle Number")
plt.ylabel("Speed (m/s)")
plt.title("Particle Speed Distribution with Mean")
plt.show()
```

Question 2

```python
#part A
import numpy as np
temperatures=np.random.normal(loc=22,scale=0.5,size=100)
temperatures
```

```python
# part B
plt.hist(temperatures,bins=20)
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.title("Distribution of Temperature Measurements")
plt.show()
```

```python
# part c
cnt=np.sum((temperatures>21.5) & (temperatures<22.5))
percentage=cnt*100/100
print(f"Percentage of measurements between 21.5°C and 22.5°C = {percentage}%")
```

Question 3

```python
# part a
velocities=np.random.exponential(scale=2.5,size=5000)
velocities
```

```python
# part B
avgs=np.array([np.average(velocities[:100]),np.average(velocities[:1000]),np.average(velocities[:3000]),np.average(velocities)])
avgs
```

```python
# part C



samples=[100,1000,3000,5000]
plt.plot(samples,avgs)
plt.axhline(np.average(velocities),linestyle="--",color="red")
plt.xlabel("Number of Samples")
plt.ylabel("Average Velocity")
plt.title("Law of Large Numbers: Convergence of Average")
```

```python
# part A
inst1=np.random.normal(loc=5.80,scale=0.15,size=500)
inst2=np.random.normal(loc=5.75,scale=0.35,size=500)
```

```python
# part B
plt.hist(inst1,bins=15,alpha=0.6,color="blue",label='Instrument 1')
plt.hist(inst2,bins=15,alpha=0.6,color="red",label='Instrument 2')
plt.xlabel("Conductivity (S/m)")
plt.ylabel("Frequency")
plt.title("Comparison of Two Instruments")
plt.legend()
```

```python
# part c
std1=np.std(inst1)
std2=np.std(inst2)
precision_ratio=std2/std1
print( f"Instrument 2 is {precision_ratio:.2f} times more scattered than Instrument 1")
```
