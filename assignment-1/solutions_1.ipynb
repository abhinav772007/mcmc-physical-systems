```python
##################PART_A##################
def cal_pressure(temp,vol):
    pressure=(8.314*temp)/(vol)
    return pressure
```

```python
##################PART_B##################
import numpy as np
import matplotlib.pyplot as plt
V = np.arange(1, 51, 1) 
T1 = 273  
T2 = 323  
P1 = [cal_pressure(T1, v) for v in V]
P2 = [cal_pressure(T2, v) for v in V]
plt.plot(V, P1, label="T = 273K")
plt.plot(V, P2, label="T = 323 K")
plt.xlabel(" V (m³)")
plt.ylabel(" P (Pa)")
plt.title("isotherms")
plt.legend()
plt.show()
```

```python
##################PART_A##################
import numpy as np
arr_10 = np.random.uniform(-100, 100, 10)
arr_100 = np.random.uniform(-100, 100, 100)
arr_1000 = np.random.uniform(-100, 100, 1000)
arr_10, arr_100, arr_1000
```

```python
##################PART_B##################
import matplotlib.pyplot as plt
plt.hist(arr_10, bins=1000, edgecolor='black')
plt.title('Histogram of Uniform Random Array (10 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.hist(arr_100, bins=1000, edgecolor='black')
plt.title('Histogram of Uniform Random Array (100 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

plt.hist(arr_1000, bins=1000, edgecolor='black')
plt.title('Histogram of Uniform Random Array (1000 elements)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

When we look at the three histograms made with 10, 100, and 1000 data points, we notice that the histogram becomes smoother as the number of points increases.

With only 10 points, the bars look very uneven. Some bins have many points while others have none, so the histogram looks rough and irregular.

With 100 points, the data is spread more evenly, so the histogram already looks better balanced.

With 1000 points, most bars are nearly the same height, so the histogram looks smooth and close to the ideal uniform distribution.

A uniform distribution means that every value between –100 and 100 has the same chance of occurring. If we had an extremely large number of points, the histogram would look almost perfectly flat.

In real situations, we only have a limited number of points, so the histogram is not perfectly flat. It is only approximately flat, with small differences in bar heights.

Randomness affects small and large datasets differently.
For small datasets, randomness has a big effect, so the histogram looks noisy.
For large datasets, random effects average out, and the histogram becomes smoother and closer to the true uniform shape.

```python
##################PART_A##################
from scipy.integrate import quad

class ThermalRod:
    def __init__(self, a=1.0, d=1.0):
        self.a = a
        self.d = d
    def _integrand(self, s):
        return 1.0 / (self.a * s + self.d)
    def get_temperature(self, x):
        resistance, _ = quad(self._integrand, 0, x)
        T_x = resistance
        return T_x

rod = ThermalRod(a=1, d=1)
T = rod.get_temperature(0.5)
T
```

```python
##################PART_B##################
import numpy as np
import matplotlib.pyplot as plt

a = 1.0
d = 1.0
L = 1.0  
rod = ThermalRod(a=a, d=d)
x_values = np.arange(0, L + 0.01, 0.01)  
T_values = [rod.get_temperature(x) for x in x_values]
plt.plot(x_values, T_values, label='T along the rod', color='blue')
plt.xlabel(' x (m)')
plt.ylabel(' T(x) (K)')
plt.title('T vs X for rod')
plt.legend()
plt.show()
```

```python
def mix_water_and_ice(M, T, m):
    L_f = 334000  
    c_water = 4180  
    Q_water = M * c_water * T  
    Q_ice = m * L_f
    if Q_water < Q_ice:
        melted_mass = Q_water / L_f
        return ('melted_mass', melted_mass)
    else:
        Q_remaining = Q_water- Q_ice
        total_mass_water = M + m  # all ice becomes water
        final_temperature = Q_remaining / (total_mass_water * c_water)
        return ('final_temperature', final_temperature)
```

```python
value = mix_water_and_ice(M=0.5, T=60, m=0.1)
print( value)
```
