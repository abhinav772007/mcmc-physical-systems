```python
# part A
import numpy as np
final_distances=[]
steps=[[1,0],[-1,0],[0,1],[0,-1]]
for i in range(1,501):
    intial=np.array([0,0])
    for j in range(1,101):
        step=steps[np.random.randint(0,4)]
        intial+=step
    r=(intial[0]**2+intial[1]**2)**0.5
    final_distances.append(r)
final_distances       


    
```

```python
# part B
import matplotlib.pyplot as plt
plt.hist(final_distances,bins=25)
plt.xlabel("Distance from Origin")
plt.ylabel("frequency")
plt.title("random walk")
plt.show()
```

```python
# part C
final_distances=np.array(final_distances)
avg_r_squared=np.mean(final_distances**2)
print(f"avgerage r_squared = {avg_r_squared}(therotical value=100)")
```

Question_2

```python
# part A
L=0.5
total=1000
cross=[]
for i in range(total):
    y=np.random.uniform(0,0.5)
    theta=np.random.uniform(0,np.pi)
    if y<=0.25*np.sin(theta):
        cross.append(1)
    else :
        cross.append(0)    
np.sum(cross)
```

```python
# part B
pi_estimate=(2*total*L)/np.sum(cross)
error=abs(pi_estimate-3.1416)/3.1416*100
print(f"Estimated Pi = {pi_estimate:4f}(true value=3.1416,error={error:4f}%")
```

```python
# part C
increments=range(100,total+1,100)
pi=[]
cross=[]
for i in range(1,total+1):
    y=np.random.uniform(0,0.5)
    theta=np.random.uniform(0,np.pi)
    if y<=0.25*np.sin(theta):
        cross.append(1)
    else :
        cross.append(0)
    if i%100==0 :
        x=(2*i*L)/np.sum(cross)
        pi.append(x)
plt.plot(list(increments),pi) 
plt.axhline(np.pi,linestyle='--',color="red")
plt.xlabel("No of Needles")
plt.ylabel("Estimated PI")
plt.title("convergence of PI Estimate via Buffon's Needle")
plt.show()               

```

```python
# part A
N_init=1000
r=0.01
U=np.random.uniform(0,1,N_init)
decay_times=-np.log(U)/r
decay_times
```

```python
# part B
time_points=np.arange(0,501,20)
remaining_nuclei=[]
analytical_N=[]
for i in time_points:
    remaining=np.sum(decay_times>i)
    remaining_nuclei.append(remaining)
    analytical_N.append(N_init*np.exp(-r*i))

remaining_nuclei,analytical_N

```

```python
# part C
plt.plot(time_points,remaining_nuclei,color="blue",label="simulation")
plt.plot(time_points,analytical_N,color="red",label="Analytical")
plt.legend()
plt.xlabel("TIme")
plt.ylabel("no of nuclei")
plt.title("Radioactive Decay: simulation X theroy")
```

```python
# part A
import itertools
dims=[1,2,3,5,10]
budget=100000
grid_m=[]
grid_evals=[]
grid_estimates=[]
grid_errors=[]
mc_estimates=[]
mc_errors=[]
for d in dims:
    exact=(1-np.exp(-1))**d
    m=int(budget**(1/d))
    grid_m.append(m)
    evals=m**d
    grid_evals.append(evals)
    points_1d=(np.arange(m)+0.5)/m
    if d<=3:
        meshes=np.meshgrid(*([points_1d]*d),indexing='ij')
        grid_points=np.stack(meshes,axis=-1).reshape(-1,d)
        f_vals=np.exp(-np.sum(grid_points,axis=1))
    else :
        f_vals=[]
        for p in itertools.product(points_1d,repeat=d):
            f_vals.append(np.exp(-np.sum(p)))
        f_vals=np.array(f_vals)
    grid_est=np.mean(f_vals)
    grid_estimates.append(grid_est)
    grid_errors.append(abs(grid_est-exact))
    x=np.random.rand(budget,d)
    f_mc=np.exp(-x.sum(axis=1))  
    mc_est=np.mean(f_mc)
    mc_estimates.append(mc_est)
    mc_errors.append(abs(mc_est-exact))      
```

```python
# part B
plt.figure()
plt.semilogy(dims,grid_evals)
plt.axhline(budget,color="red")
plt.xlabel("dim d")
plt.ylabel("no of grid evaluations")
plt.title("exponential growth of evaluations")
```

```python
plt.semilogy(dims, grid_errors,label="Grid error")
plt.semilogy(dims, mc_errors,label="Monte Carlo error")
plt.xlabel("Dimension d")
plt.ylabel("Absolute Error")
plt.title("Grid vs Monte Carlo Error")
plt.show()
```

1.The grid-based integration method fails in high dimensions because the number of grid points grows exponentially with dimension, quickly exceeds fixed computational budget. As a result, the grid becomes extremely coarse in high dimensions, leading to large approximation errors—this is known as the curse of dimensionality.<br>2.Monte Carlo integration succeeds because its convergence rate depends primarily on the no of samples and is largely independent of dimension. Even in high dimensions, Monte Carlo maintains a stable error as long as enough random samples are used.<br>3.In practice, these methods are far more suitable for high-dimensional integrals comes in physics, finance, and machine learning.
